## Data files and summary statistics

```python
import csv # para leer csv
# tener el archivo csv en la misma carpeta que el archivo de python
%precision 2 #tener 2 decimales

with open ('mpg.csv')as csvfile:
    mpg=list(csv.DictReader(csvfile))
mpg[:3]
```

```python
len(mpg)
```

```python
# para ver los nombres de las variables
mpg[0].keys()
```

```python
#para hacer operaciones tener cuidado con los tipos de datos
# supongamos queremos determinar las millas por galon promedio en cuidad 
# en todos los vehiculos del archivo del archivo csv

sum(float(d["cty"]) for d in mpg)/len(mpg)
```

```python
sum(float(d["hwy"]) for d in mpg)/len(mpg)
```

```python
cylinders=set(d["cyl"]for d in mpg)
cylinders
```

```python
CtyMpgByCyl=[]

for c in cylinders:
    summpg=0
    cyltypecount=0
    for d in mpg:
        if d["cyl"]==c:
            summpg+=float(d["cty"])
            cyltypecount+=1
    CtyMpgByCyl.append((c,summpg/cyltypecount))
    
CtyMpgByCyl.sort(key=lambda x:x[0])
CtyMpgByCyl
                 
```

```python
vehicleclasses=set(d["class"] for d in mpg)
vehicleclasses
```

```python
HwyMpgByClass=[]

for t in vehicleclasses:
    summpg=0
    vclasscount=0
    for d in mpg:
        if d["class"]==t:
            summpg+=float(d["hwy"])
            vclasscount+=1
    HwyMpgByClass.append((t,summpg/vclasscount))
# vamos a ordenar de menor a mayor los resultados
HwyMpgByClass.sort(key=lambda x:x[1])
HwyMpgByClass

    
```

# The Python Programming Languages: Objetcs and maps()

```python
# ejemplo de clase, se tiene public y private en el mismo lugar
class Person:
    department="school of information"
    
    def set_name(self, new_name):
        self.name=new_name
    def set_location(self, new_location):
        self.location=new_location
   
```

```python
person=Person()
person.set_name("Luis Amagua")
person.set_location("Quito")

print("{} live in {} and work in de department {}".format(person.name, person.location, person.department) )
```

vamos a usar la funcion map

```python
store1=[10.00, 11.00, 12.34, 2.34]
store2=[9.00, 11.10, 12.34, 2.01]

# quiero encontrar el valor minimo de varios items en las tiendas
cheapest=map(min, store1,store2)
print(cheapest) # solamente me devuelve al referencia, no el valor

# si queremos ver el resultado 
for item in cheapest:
    print(item)
```

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

# ¿como puedo retornar el apellido y el titulo (sin el nombre)?

def split_title_and_name(person):
    return person.split(" ")[0]+person.split(" ")[2]

split_title_and_name('Dr. Christopher Brooks')
list(map(split_title_and_name,people))
```

```python
# una alternaiva a lo anterior, es tomar los valores de titulo y apellido de forma separada
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    title=person.split(" ")[0]
    lastname=person.split(" ")[2]
    return "{} {}".format(title, lastname)

list(map(split_title_and_name,people))
```

```python
def times_tables():
    lst = []
    for i in range(10):
        for j in range (10):
            lst.append(i*j)
    return lst

# por comprension es mas util
times_tables() == [j*i for i in range(10) for j in range(10)]
```
